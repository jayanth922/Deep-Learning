## Task 1 – Inference with a pretrained sentiment-classifier (Keras-NLP)

```python
# Imports + verbose logging setup
import os, logging, tensorflow as tf, keras_nlp

# logging switches
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "0"
tf.get_logger().setLevel(logging.DEBUG)
tf.debugging.set_log_device_placement(True)

# Load a ready-made SST-2 classifier preset
PRESET = "bert_tiny_en_uncased_sst2"
classifier = keras_nlp.models.BertClassifier.from_preset(PRESET)

classifier.summary()          # architecture + param counts

# Run inference on example sentences
texts = [
    "I absolutely loved this movie—fantastic performances!",
    "It was a dull, boring film. I walked out halfway."
]

logits  = classifier.predict(texts,
                             batch_size=2,
                             verbose=2)

probs   = tf.nn.softmax(logits, axis=-1)   # convert to probabilities
labels  = tf.argmax(probs, axis=-1).numpy()  # 0 = negative, 1 = positive

# Pretty-print results with probabilities
for t, p, lab in zip(texts, probs.numpy(), labels):
    sentiment = "positive" if lab == 1 else "negative"
    print(f"{t[:60]:<62} → {sentiment:8}  "
          f"(neg={p[0]:.3f}, pos={p[1]:.3f})")
```

## Task 2 - DistilBERT on 2 k IMDB examples

```python
import os, tensorflow as tf
from datasets import load_dataset
from transformers import (
    DistilBertTokenizerFast,
    TFDistilBertForSequenceClassification,
    DataCollatorWithPadding,
)

# logging
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

MODEL   = "distilbert-base-uncased"
MAX_LEN = 96
BATCH   = 16
EPOCHS  = 1
N_SAMP  = 2_000           # tiny subset ⇒ quick run

tok   = DistilBertTokenizerFast.from_pretrained(MODEL)
model = TFDistilBertForSequenceClassification.from_pretrained(
    MODEL, num_labels=2
)

# data
imdb  = load_dataset("imdb")
train = (
    imdb["train"].shuffle(42).select(range(N_SAMP))
    .map(lambda b: tok(b["text"], truncation=True, max_length=MAX_LEN),
         batched=True, remove_columns=["text"])
    .rename_column("label", "labels")
)
test = (
    imdb["test"].shuffle(42).select(range(N_SAMP))
    .map(lambda b: tok(b["text"], truncation=True, max_length=MAX_LEN),
         batched=True, remove_columns=["text"])
    .rename_column("label", "labels")
)

collator = DataCollatorWithPadding(tok, return_tensors="tf")

tf_train = train.to_tf_dataset(
    columns=["input_ids", "attention_mask"],
    label_cols=["labels"],
    shuffle=True,
    batch_size=BATCH,
    collate_fn=collator,
)
tf_val = test.to_tf_dataset(
    columns=["input_ids", "attention_mask"],
    label_cols=["labels"],
    shuffle=False,
    batch_size=BATCH,
    collate_fn=collator,
)

# compile & train
model.compile(
    optimizer="adam",
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"],
)

history = model.fit(
    tf_train,
    validation_data=tf_val,
    epochs=EPOCHS,
    verbose=2,
)

# inference
samples = [
    "Wonderful film; I was moved to tears!",
    "Utter waste of time. Boring and predictable."
]
inputs  = tok(samples, return_tensors="tf", padding=True,
              truncation=True, max_length=MAX_LEN)
logits  = model(inputs).logits
probs   = tf.nn.softmax(logits, axis=-1).numpy()

print("\nPredictions:")
for t, p in zip(samples, probs):
    print(f"  {t[:55]:<55} → neg={p[0]:.3f}  pos={p[1]:.3f}")
```

## Task 3 – Build & train a tiny Transformer from scratch

```python
import os, tensorflow as tf
from tensorflow.keras import layers
from datasets import load_dataset

# logging
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

# Hyper-params (tiny → fast demo)
VOCAB_SIZE, SEQ_LEN   = 20_000, 96
EMBED_DIM, NUM_HEADS  = 64, 2
FF_DIM, NUM_LAYERS    = 128, 2
BATCH, EPOCHS, N_SAMP = 32, 3, 2_000

# Data
imdb       = load_dataset("imdb")
train_raw  = imdb["train"].shuffle(42).select(range(N_SAMP))
test_raw   = imdb["test"] .shuffle(42).select(range(N_SAMP))

vectorizer = layers.TextVectorization(
    max_tokens=VOCAB_SIZE,
    output_mode="int",
    output_sequence_length=SEQ_LEN,
    standardize="lower_and_strip_punctuation",
)
vectorizer.adapt(train_raw["text"])

def preprocess(text, label):
    x = tf.expand_dims(text, -1)
    x = vectorizer(x)
    x = tf.squeeze(x, axis=0)
    return tf.cast(x, tf.int32), tf.cast(label, tf.int32)

tf_train = (
    tf.data.Dataset.from_tensor_slices((train_raw["text"], train_raw["label"]))
    .map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    .shuffle(10_000).batch(BATCH).prefetch(tf.data.AUTOTUNE)
)
tf_val = (
    tf.data.Dataset.from_tensor_slices((test_raw["text"], test_raw["label"]))
    .map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(BATCH).prefetch(tf.data.AUTOTUNE)
)

# Transformer blocks
class PositionalEmbedding(layers.Layer):
    def __init__(self, vocab_size, embed_dim, seq_len):
        super().__init__()
        self.token_embed = layers.Embedding(vocab_size, embed_dim)
        self.pos_embed   = layers.Embedding(seq_len, embed_dim)
    def call(self, x):
        positions = tf.range(tf.shape(x)[1])
        return self.token_embed(x) + self.pos_embed(positions)

def transformer_block(x, embed_dim, num_heads, ff_dim, rate=0.1):
    attn_out = layers.MultiHeadAttention(num_heads, key_dim=embed_dim)(x, x)
    x = layers.LayerNormalization(epsilon=1e-6)(x + layers.Dropout(rate)(attn_out))
    ffn = layers.Dense(ff_dim, activation="relu")(x)
    ffn = layers.Dense(embed_dim)(ffn)
    return layers.LayerNormalization(epsilon=1e-6)(x + layers.Dropout(rate)(ffn))

# Build the model
inputs = layers.Input(shape=(SEQ_LEN,), dtype=tf.int32)
x = PositionalEmbedding(VOCAB_SIZE, EMBED_DIM, SEQ_LEN)(inputs)
for _ in range(NUM_LAYERS):
    x = transformer_block(x, EMBED_DIM, NUM_HEADS, FF_DIM)
x = layers.GlobalAveragePooling1D()(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
model.summary(line_length=120)

# Train
history = model.fit(tf_train,
                    validation_data=tf_val,
                    epochs=EPOCHS,
                    verbose=2)

# Quick demo predictions
samples = [
    "Wonderful film; I was moved to tears!",
    "Utter waste of time. Boring and predictable."
]
token_ids = vectorizer(samples)
probs     = model.predict(token_ids, verbose=0).flatten()

print("\nPredictions:")
for s, p in zip(samples, probs):
    print(f"  {s[:55]:<55} → positive_prob={p:.3f}")
```
